```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('../datasets/rank_1/lyric_sentiment.csv', na_values=0.0)
pd.set_option('display.max_columns', None)
df.head()
```

```python
float_cols = df.select_dtypes(include=['float']).columns
df[float_cols] = df[float_cols].fillna(df[float_cols].mean())

scaled_df = df.copy()

scaler = MinMaxScaler(feature_range=(0, 1))
scaled_df[float_cols] = scaler.fit_transform(df[float_cols])
```

```python
scaled_df.head()
```

```python
scaled_df.to_csv('../datasets/rank_1/lyric_sentiment_scaled.csv', index=False)
```
