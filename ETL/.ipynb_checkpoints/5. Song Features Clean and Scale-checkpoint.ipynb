```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('../datasets/features_and_analysis.csv')

df['duration'] = df['duration_ms'].apply(lambda x: f"{x // 60000}:{(x % 60000) // 1000:02d}")

df = df.drop(columns=['duration_ms'])

cols_to_scale = ['danceability', 'energy', 'speechiness', 'loudness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo',
                 'tempo_confidence', 'time_signature_confidence', 'key_confidence', 'mode_confidence']

df['instrumentalness'] = df['instrumentalness'] * 1e6
df['loudness'] = df['loudness'] + 50

scaler = MinMaxScaler(feature_range=(-1, 1))

for col in cols_to_scale:
    if col not in ['tempo', 'loudness']:
        col_log = np.log1p(df[col].values)
    df[col] = scaler.fit_transform(col_log.reshape(-1, 1))

df.to_csv('../datasets/features_and_analysis_scaled.csv', index=False)

pd.set_option('display.max_columns', None)
df.head()
```
