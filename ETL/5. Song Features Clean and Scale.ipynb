```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('../datasets/rank_1/features.csv')
pd.set_option('display.max_columns', None)
df.head()
```

```python
df = df.drop(columns=['duration_ms', 'key', 'mode', 'time_signature'])

df['instrumentalness'] = df['instrumentalness'] * 1e6
df['loudness'] = df['loudness'] + 50

df.head()
```

```python
cols_to_scale = ['danceability', 'energy', 'speechiness', 'loudness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo',
                 'duration', 'tempo_confidence', 'time_signature_confidence', 'key_confidence', 'mode_confidence']
```

```python
scaler = MinMaxScaler(feature_range=(-1, 1))

for col in cols_to_scale:
    if col not in ['tempo', 'loudness']:
        col_log = np.log1p(df[col].values)
    df[col] = scaler.fit_transform(col_log.reshape(-1, 1))
```

```python
df.head()
```

```python
df.to_csv('datasets/rank_1/features_scaled.csv', index=False)
```
