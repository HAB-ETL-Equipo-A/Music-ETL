```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar los archivos CSV
top_1_songs = pd.read_csv('complete_filled_top_1_songs.csv')
random_songs = pd.read_csv('canciones_aleatorias_con_fechas_y_datos.csv')

key_map = {
    0: 'C', 1: 'C#', 2: 'D', 3: 'D#', 4: 'E',
    5: 'F', 6: 'F#', 7: 'G', 8: 'G#', 9: 'A',
    10: 'A#', 11: 'B'
}

def get_key_name(row):
    key_name = key_map.get(row['key'])
    if row['mode'] == 1:  # Mayor
        return f"{key_name} Maj"
    else:  # Menor
        return f"{key_name} Min"

top_1_songs['tonalidad'] = top_1_songs.apply(get_key_name, axis=1)

random_songs['tonalidad'] = random_songs.apply(
    lambda row: f"{key_map.get(row['key'], 'Unknown')} Maj" if row['mode'] == 1 else f"{key_map.get(row['key'], 'Unknown')} Min",
    axis=1
)

key_counts_top_1 = top_1_songs['tonalidad'].value_counts().reset_index()
key_counts_top_1.columns = ['Tonalidad', 'Frecuencia']
key_counts_top_1 = key_counts_top_1.sort_values(by='Frecuencia', ascending=False)

key_counts_random = random_songs['tonalidad'].value_counts().reset_index()
key_counts_random.columns = ['Tonalidad', 'Frecuencia']
key_counts_random = key_counts_random.sort_values(by='Frecuencia', ascending=False)

key_counts_top_1['Fuente'] = 'Top 1 Songs'
key_counts_random['Fuente'] = 'Random Songs'

combined = pd.concat([key_counts_top_1[['Tonalidad', 'Frecuencia', 'Fuente']], key_counts_random[['Tonalidad', 'Frecuencia', 'Fuente']]])

sns.set(style="whitegrid")

plt.figure(figsize=(12, 8))
sns.barplot(data=combined, y='Tonalidad', x='Frecuencia', hue='Fuente', dodge=True, palette='Set2')

plt.title('Frecuencia de canciones por tonalidad (Top 1 vs Random Songs)')
plt.xlabel('Frecuencia')
plt.ylabel('Tonalidad')
plt.legend(title='Fuente')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.preprocessing import StandardScaler

file_path = 'complete_filled_top_1_songs.csv'
data = pd.read_csv(file_path)

fecha_columna = 'date'
if data[fecha_columna].dtype != 'datetime64[ns]':
    data[fecha_columna] = pd.to_datetime(data[fecha_columna], errors='coerce')

data['year'] = data[fecha_columna].dt.year

columnas_numericas = data.select_dtypes(include=['float64', 'int64']).columns

data_by_year = data.groupby('year')[columnas_numericas].mean().reset_index()

scaler = StandardScaler()
data_normalized = pd.DataFrame(scaler.fit_transform(data_by_year[columnas_numericas]), columns=columnas_numericas)
data_normalized['year'] = data_by_year['year']

plt.figure(figsize=(14, 8))

def plot_polynomial_fit(x, y, degree=3, label=None, color='blue', linewidth=3):
    # Ajuste polinómico
    coeffs = np.polyfit(x, y, degree)
    p = np.poly1d(coeffs)
    x_fit = np.linspace(x.min(), x.max(), 100)
    y_fit = p(x_fit)

    plt.plot(x_fit, y_fit, label=label, color=color, linewidth=linewidth)

plot_polynomial_fit(data_normalized['year'], data_normalized['tempo'], label='Tempo', color='blue', linewidth=3)
plot_polynomial_fit(data_normalized['year'], data_normalized['danceability'], label='Danceability', color='orange', linewidth=3)
plot_polynomial_fit(data_normalized['year'], data_normalized['energy'], label='Energy', color='green', linewidth=3)
plot_polynomial_fit(data_normalized['year'], data_normalized['loudness'], label='Loudness', color='red', linewidth=3)

plt.title('Tempo, Danceability, Energy y Loudness de los Top 1 a lo largo del Tiempo', fontsize=16)
plt.xlabel('Año', fontsize=14)
plt.ylabel('Valor Normalizado', fontsize=14)
plt.legend(title='Características', fontsize=12)
plt.grid(True)
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Este gráfico muestra la evolución normalizada de cuatro características 
# musicales de las canciones top 1 en Billboard desde 1958 hasta el presente: 
# Tempo, Danceability, Energy y Loudness.

# Análisis de las tendencias:

# 1. Tempo (azul):
# - El tempo disminuyó significativamente entre 1958 y mediados de los años 70, 
# alcanzando un valor mínimo alrededor de los 1980, pero luego comenzó a 
# incrementarse gradualmente hasta los tiempos modernos.
# - Motivo potencial:
# - En los años 60 y 70, los géneros dominantes como el rock psicodélico, el 
# folk y el soul tenían tempos más lentos para reflejar una mayor introspección 
# y experimentación.
# - Desde los años 2000, el crecimiento de la música electrónica, hip-hop y pop 
# con ritmos más rápidos y bailables podría explicar el aumento.

# 2. Danceability (naranja):
# - La bailabilidad comienza siendo baja, pero aumenta significativamente desde 
# los años 80 hasta el presente.
# - Motivo potencial:
# - En los 80, con la popularización de géneros como el disco, el synth-pop y el 
# dance music, la música bailable se convirtió en el foco principal de las 
# pistas de baile y en los éxitos de Billboard.
# - En décadas más recientes, el pop y el reguetón, con estructuras rítmicas más 
# enfocadas en la bailabilidad, explican el aumento.

# 3. Energy (verde):
# - La energía aumenta gradualmente desde 1958, alcanzando un máximo alrededor 
# del 2000, pero parece haber comenzado a estabilizarse o disminuir ligeramente 
# en los años recientes.
# - Motivo potencial:
# - La música de los 60 y 70 tenía una energía más baja debido a los géneros 
# dominantes (folk, baladas y rock más suave).
# - El aumento desde los 80 puede atribuirse a la introducción de géneros más 
# intensos como el hard rock, el punk, el EDM, y eventualmente el hip-hop.
# - El ligero descenso reciente podría reflejar el auge de canciones más 
# minimalistas y melódicas en géneros como el pop alternativo y el indie.

# 4. Loudness (rojo):
# - El volumen aumentó drásticamente desde los años 70, alcanzando un máximo en 
# los 2000, con una ligera estabilización en años recientes.
# - Motivo potencial:
# - Esto está relacionado con la loudness war (guerra del volumen), donde las 
# canciones se producían cada vez más fuertes para captar la atención del 
# oyente.
# - La estabilización reciente puede reflejar una mayor conciencia de la calidad 
# del sonido y de la dinámica en la producción musical.

# Conclusionesssssss
# - Cambio en la música dominante: Los años 80 marcaron un cambio hacia música 
# más energética, bailable y rítmica, reflejando la influencia del disco, pop y 
# el dance.
# - Auge de la tecnología en la música: La producción musical moderna (samples, 
# autotune, mastering avanzado) permitió aumentar la energía y el volumen de 
# las canciones sin perder claridad.
# - Diversidad y globalización reciente: En años recientes, la influencia de 
# géneros globales como el reguetón, el K-pop y el afrobeats ha favorecido 
# canciones más bailables, con tempos variados y volúmenes controlados.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

file_path = 'complete_filled_top_1_songs.csv'
data = pd.read_csv(file_path)

columns = ['tempo', 'danceability', 'energy', 'loudness']
data = data[columns]

sns.set(style="whitegrid")

feature_pairs = [
    ('tempo', 'danceability'),
    ('tempo', 'energy'),
    ('tempo', 'loudness'),
    ('danceability', 'energy'),
    ('danceability', 'loudness'),
    ('energy', 'loudness')
]

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()

for i, (feature1, feature2) in enumerate(feature_pairs):
    sns.regplot(x=feature1, y=feature2, data=data, scatter_kws={'s': 20, 'alpha': 0.1}, line_kws={'color': 'red'}, ax=axes[i])
    axes[i].set_title(f'{feature1} vs {feature2}')
    axes[i].set_xlabel(feature1)
    axes[i].set_ylabel(feature2)

plt.tight_layout()

plt.show()
```

```python
# Se pueden observar varias relaciones entre las variables
# relacionadas con las canciones top 1 en Billboard
# desde 1958. Estas son algunas conclusiones generales
# que se pueden extraer de los gráficos:

# Tempo vs Danceability (bailabilidad):

#Parece haber una ligera relación negativa: a medida que aumenta el tempo
# (velocidad de la canción), la bailabilidad tiende a disminuir. Esto podría 
# indicar que canciones más rápidas no necesariamente son más bailables.
# Tempo vs Energy:

# Existe una ligera relación positiva: canciones con tempos más altos tienden
# a tener una mayor energía. Esto tiene sentido porque canciones más 
# rápidas suelen ser percibidas como más dinámicas.

# Tempo vs Loudness (volumen):

# La correlación entre el tempo y el volumen parece ser débil, con una línea 
# de tendencia casi plana. Esto sugiere que el tempo no está directamente
# relacionado con lo fuerte que suena una canción.

# Danceability vs Energy:

# Hay una relación positiva clara: canciones que son más bailables 
# también tienden a ser más enérgicas. Esto podría reflejar que
# canciones ideales para bailar suelen tener ritmos vibrantes y dinámicos.

# Danceability vs Loudness:

# Existe una ligera relación positiva: canciones más bailables
# tienden a ser más fuertes. Esto tiene sentido porque la
# música para bailar generalmente se produce con un volumen más alto.

# Energy vs Loudness:

# La relación es marcadamente positiva: canciones con mayor energía
# tienden a ser más fuertes. Esto es esperado, ya que el nivel de 
# energía de una canción suele reflejarse en el volumen y la intensidad de la mezcla.

# Observaciones Generales:
# Las relaciones más fuertes parecen ser entre Energy y Loudness y entre Danceability
# y Energy, lo cual es coherente con lo que se esperaría en música popular.
# Algunas variables, como el tempo, tienen una relación más débil con otras
# como la bailabilidad o el volumen, lo que sugiere que el ritmo no es el único
# determinante de estas cualidades.
```
